```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rc('text', usetex=True)
plt.rc('font', family='serif')
```

```python
df = pd.read_csv(f'testActiveDiffusion/log.dat', sep=' ')
# df.head()
fig, ax = plt.subplots(1, 1, figsize=(10,5))
## Theory, infinite D_T
ax.plot(df['i'], df['H'], '-', color='#fdae61', linewidth=4.0, label=r'Simulations t = 10')

ax.set_xlabel(r'$R-r_0$', fontsize=20)
ax.set_ylabel(r'$c$', fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=15)
# log scale
ax.set_yscale('log')
ax.set_xscale('log')

ax.set_xlim([1e0, df['i'].max()])
ax.set_ylim([1e-3, 1e-1])

ax.legend(loc='best')

# plt.savefig('test.pdf', bbox_inches='tight')
plt.show()
```

```python
t = 0.
df1 = pd.read_csv(f'testActiveDiffusion/intermediate/fieldData-{t:5.4f}.dat', sep=' ')
t = 10.
df2 = pd.read_csv(f'testActiveDiffusion/intermediate/fieldData-{t:5.4f}.dat', sep=' ')

c0, A, L0 = 0., 1., 8.0

fig, ax = plt.subplots(1, 1, figsize=(10,5))
rout = 0.999*L0

## Theory, infinite D_T
y2 = np.linspace(1.0, rout, 10000)
cL = c0 - A*np.log(y2/rout)

ax.plot(df1['y']-1., df1['cL'], '-', color='k', linewidth=4.0, label=r'Simulations t = 0')
ax.plot(df2['y']-1., df2['cL'], '-', color='#fdae61', linewidth=4.0, label=r'Simulations t = 10')
ax.plot(y2-1, cL, '--', color='gray', linewidth=4.0, label=r'Theory: tracer, infinite $D_T$')

ax.set_xlabel(r'$R-r_0$', fontsize=20)
ax.set_ylabel(r'$c$', fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=15)
# log scale
ax.set_yscale('log')
ax.set_xscale('log')

ax.set_xlim([1e-3, rout])
ax.set_ylim([1e-3, 4.0])

ax.legend(loc='best')

# plt.savefig('test.pdf', bbox_inches='tight')
plt.show()
```
